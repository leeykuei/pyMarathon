```python
import pandas as pd
```

```python
#讀取STOCK_DAY_0050_202009.csv、STOCK_DAY_0050_202010.csv

stock_data1 = pd.read_csv('STOCK_DAY_0050_202009.csv')
stock_data2 = pd.read_csv('STOCK_DAY_0050_202010.csv')
```

```python
#串聯
stock_data =pd.concat([stock_data1,stock_data2],axis=0)
stock_data.reset_index(inplace=True)
stock_data
```

```python
#找出open小於close的資料
stock_data.loc[stock_data.open < stock_data.close]
```

```python
'''3. 請根據提供的資料，選擇出下列的要求：
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100

'''
```

```python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
```

```python
# - 1. filtered by first column > 20?
filter = df[0] >20
df.loc[filter]
```

```python
# - 2. filtered by first column + second column > 50
filter = (df[0]+ df[1] ) > 50
df.loc[filter]
```

```python
# - 3. filtered by first column < 30 or second column > 30
filter = (df[0] < 30) |  (df[1]>30)
df.loc[filter]
```

```python
# - 4. filtered by total sum of row > 100
filter = df.sum(axis=1) > 100
df[filter]
```
