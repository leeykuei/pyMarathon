```python
import seaborn as sns
import numpy as np
```

```python
#1.繪製模型殘差 圖型
#設定圖形樣式 - whitegrid
# use sns.set
sns.set(style='whitegrid')
# 利用 NUMPY 去建立資料集
# np.random.RandomState 設定數學式
rs = np.random.RandomState(7)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)

# 畫圖
sns.residplot(x, y, lowess=True, color="b")
```

```python
#2.使用 distplot()使用簡單的規則來正確猜測預設情況下正確的數位,但嘗試更多或更少的 bin 可能會顯示資料中的其他特徵：
#有無kde對圖形分布的影響

sns.set(color_codes=True)

# bin: 指的是特徵值, 
# kde: on/off
# sns.distplot();

sns.distplot(x, bins=15, kde=False, rug=True);
```

```python
sns.distplot(x, bins=15, kde=True, rug=True);
```

```python
sns.distplot(x, bins=30, kde=True, rug=True);
```
